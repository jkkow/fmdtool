```python
import sys
sys.path.append('C:/Users/jkkow/repo/')
from fmdtool.mtool import LPModes
from fmdtool.viewer import ModeShow
import matplotlib.pyplot as plt
import numpy as np
```

```python
lps = LPModes(v = 7.5, d=10)
# Sort the uset items by value in accending order
sorted_uset = sorted(lps.uset.items(), key=lambda x: x[1])
for key, value in sorted_uset:
    print(f"{key}: {value:.4f}")
print(sorted_uset)
```

```python
ms = ModeShow(lps)
lp11 = lps.LP(1, 1, rot=90, jones=(1/np.sqrt(2), 1j/np.sqrt(2)))

fig, axe = plt.subplots()
axe.set_aspect("equal")
ms.plot2d(fig, axe, lp11, core='on', tickoff=True, vector='on')
```

```python
# generates mode object of possible mode set
mset = {}
mset['lp01'] = lps.LP(0, 1)
mset['lp11a'] = lps.LP(1, 1, 90)
mset['lp11b'] = lps.LP(1, 1)
mset['lp21a'] = lps.LP(2, 1)
mset['lp21b'] = lps.LP(2, 1, 45)
mset['lp02'] = lps.LP(0, 2)
mset['lp31a'] = lps.LP(3, 1, 30)
mset['lp31b'] = lps.LP(3, 1)
mset['lp12a'] = lps.LP(1, 2, 90)
mset['lp12b'] = lps.LP(1, 2)
mset['lp41a'] = lps.LP(4, 1)
mset['lp41b'] = lps.LP(4, 1, 22.5)
mset['lp22a'] = lps.LP(2, 2)
mset['lp22b'] = lps.LP(2, 2, 45)
mset['lp03'] = lps.LP(0, 3)
```

```python
ms = ModeShow(lps)
fig, axes = plt.subplots(nrows=5, 
                         ncols=3, 
                         figsize=(12, 20),
                         gridspec_kw={
    'width_ratios': [1, 1, 1], # ratio of column with
    'height_ratios': [1, 1, 1, 1, 1]
                         })
# fig.subplots_adjust(hspace=0, wspace=0)

axes_flat = axes.flatten() # flattening ndarray
for ax, lp in zip(axes_flat, mset.keys()):
    ms.plot2d(fig, ax, mset[lp], 
              tickoff = True,
              core='on',
              vector='on')
    ax.set_title(lp.upper())
```
